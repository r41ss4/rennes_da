```python
# Review the data and how many rows have missing information in each columns
data_f1.isna().sum()
```

```python
# Find rows with null values in the Route and Total_Stops columns.
null_rows = data_f1[data_f1['Route'].isnull() | data_f1['Total_Stops'].isnull()]
print(null_rows)
```

```python
# Delete rows with null values in the Route and Total_Stops columns.
data_f1 = data_f1.dropna(subset=['Route', 'Total_Stops'])
```

```python
# Review the data and how many rows have missing information in each columns
data_f1.isna().sum()
```

```python
# check duplicated data
duplicate_rows = data_f1[data_f1.duplicated()]
print(duplicate_rows)
```

```python
# Delete duplicate rows, keeping the first occurrence
data_f1 = data_f1.drop_duplicates()
```
