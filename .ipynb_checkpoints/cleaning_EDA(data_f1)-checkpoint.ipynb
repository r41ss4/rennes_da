```python
# Convert Dep_Time and Arrival_Time to datetime, then extract the hour
data_f1['Dep_Hour'] = pd.to_datetime(data_f1['Dep_Time'], format='%H:%M').dt.hour
data_f1['Arrival_Hour'] = pd.to_datetime(data_f1['Arrival_Time'], format='%H:%M', errors='coerce').dt.hour

def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

data_f1['Dep_Time_Category'] = data_f1['Dep_Hour'].apply(time_of_day)
data_f1['Arrival_Time_Category'] = data_f1['Arrival_Hour'].apply(time_of_day)

dep_price_summary = data_f1.groupby('Dep_Time_Category')['Price'].mean().reset_index()
arrival_price_summary = data_f1.groupby('Arrival_Time_Category')['Price'].mean().reset_index()
dep_price_summary
```

```python
arrival_price_summary
```

```python
# How the Ticket Price change based on the Departure Time?
plt.figure(figsize=(4, 3))
order = ['Morning', 'Afternoon', 'Evening', 'Night']
sns.barplot(x='Dep_Time_Category', y='Price', data=dep_price_summary, palette='viridis', order=order)
plt.title('Average Ticket Price by Departure Time Category',fontsize=9)
plt.xlabel('Departure Time Category',fontsize=8)
plt.ylabel('Average Price',fontsize=8)
plt.xticks(fontsize=8)
plt.show()
```

```python
# How the Ticket Price change based on the Arrival Time?
plt.figure(figsize=(4, 3))
sns.barplot(x='Arrival_Time_Category', y='Price', data=arrival_price_summary, palette='viridis', order=order)
plt.title('Average Ticket Price by Arrival Time Category',fontsize=9)
plt.xlabel('Arrival Time Category',fontsize=8)
plt.ylabel('Average Price',fontsize=8)
plt.xticks(fontsize=8)
plt.show()
```

```python
# Convert Duration to total minutes
def duration_to_minutes(duration):
    hours = 0
    minutes = 0
    if 'h' in duration:
        hours = int(duration.split('h')[0])
        duration = duration.split('h')[1]
    if 'm' in duration:
        minutes = int(duration.split('m')[0])
    return hours * 60 + minutes
```

```python
# Apply conversion
data_f1['Duration_Minutes'] = data_f1['Duration'].apply(duration_to_minutes)
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Duration_Minutes', y='Price', data=data_f1, alpha=0.5)
plt.title('Flight Duration vs. Ticket Price')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
# Correlation Analysis
correlation = data_f1['Duration_Minutes'].corr(data_f1['Price'])
print(f"Correlation between Duration and Price: {correlation:.2f}")
```
